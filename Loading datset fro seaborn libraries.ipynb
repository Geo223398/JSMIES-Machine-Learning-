```python
import pandas as pd
import  numpy as np
import seaborn as sns
import matplotlib.pylab as plt
from pydataset import data
import warnings
warnings.filterwarnings('ignore')
```

#### Loading data set from seaborn library

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('car_crashes')
df
```

```python
df.info()
```

```python
df.abbrev.nunique()
```

```python
sns.heatmap(data=df.corr(),annot=True)
plt.show()
```

```python
sns.lineplot(data=df,x='alcohol',y='speeding')
plt.axhline(df.speeding.mean(),color="g")
plt.show()
```

```python
sns.barplot(data=df[['ins_premium','ins_losses']])
plt.show()
```

```python
sns.scatterplot(data=df,x='alcohol',y='ins_losses', hue='alcohol')
plt.axhline(df.ins_losses.mean(),color='g')
plt.show()
```

```python
df["losed_Amound"]=df['ins_premium']-df['ins_losses']
```

```python
df
```

```python
sns.lineplot(data=df,x='alcohol',y='losed_Amound')
plt.show()
```
